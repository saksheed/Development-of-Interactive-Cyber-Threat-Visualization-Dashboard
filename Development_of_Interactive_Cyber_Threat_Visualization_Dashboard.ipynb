This task will display the frequency distribution of different attack types.

### Task 5: Identify Unique Values in Key Categorical Columns

### Task 6: Analyze Severity Distribution
This task examines how cyber incidents are distributed across severity levels.

```python
print('Distribution of severity:')
display(df['severity'].value_counts())
```

### Task 7: Standardize Text Formatting
This task ensures consistent capitalization across key categorical columns by converting their values to lowercase.

```python
categorical_columns = ['attack_type', 'attack_device', 'security_type', 'severity']
for col in categorical_columns:
    df[col] = df[col].str.lower()
    print(f"Column '{col}' standardized to lowercase.")

print("\nFirst 5 rows after standardization:")
display(df.head())
```

### Task 8: Validate IP Address Format
This task checks whether attacker and target IP addresses follow valid IPv4 format.

```python
import re

def is_valid_ip(ip):
    pattern = r'^(\d{1,3}\.){3}\d{1,3}$'
    return bool(re.match(pattern, ip))

df['valid_attacker_ip'] = df['attacker_ip'].apply(is_valid_ip)
df['valid_target_ip'] = df['target_ip'].apply(is_valid_ip)

display(df[['attacker_ip', 'valid_attacker_ip', 'target_ip', 'valid_target_ip']].head())
```

### Task 9: Detect Outliers in Attack Distribution
This task identifies attack types with unusually high or low occurrence.

```python
attack_counts = df['attack_type'].value_counts()

mean_count = attack_counts.mean()
std_count = attack_counts.std()

outliers = attack_counts[(attack_counts > mean_count + std_count) |
                          (attack_counts < mean_count - std_count)]

display(outliers)
```

### Task 10: Analyze Security Controls vs Severity
This task evaluates how different security mechanisms relate to incident severity.

```python
security_severity = pd.crosstab(df['security_type'], df['severity'])
display(security_severity)
```

### Task 11: Visualize Security Controls vs Severity

This task visualizes the relationship between different security mechanisms and incident severity using a heatmap to identify patterns and concentrations.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.heatmap(security_severity, annot=True, fmt='d', cmap='YlGnBu', linewidths=.5)
plt.title('Security Controls vs. Incident Severity')
plt.xlabel('Severity')
plt.ylabel('Security Type')
plt.show()
```

### Task 12: Visualize Distribution of Attack Types
This graph shows how frequently each cyber attack type occurs.

```python
import matplotlib.pyplot as plt

attack_counts = df['attack_type'].value_counts()

plt.figure(figsize=(10, 6))
attack_counts.plot(kind='bar')
plt.xlabel('Attack Type')
plt.ylabel('Number of Incidents')
plt.title('Distribution of Attack Types')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Task 13: Visualize Severity Distribution
This graph displays the distribution of incidents across severity levels.

```python
import matplotlib.pyplot as plt

severity_counts = df['severity'].value_counts()

plt.figure(figsize=(8, 5))
severity_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Severity Level')
plt.ylabel('Number of Incidents')
plt.title('Severity Distribution of Cyber Attacks')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Task 14: Visualize Attacks by Attacker Location
This graph highlights the top 10 countries by number of attacks.

```python
import matplotlib.pyplot as plt

location_counts = df['attacker_location'].value_counts().head(10)

plt.figure(figsize=(12, 6))
location_counts.plot(kind='bar', color='salmon')
plt.xlabel('Attacker Location')
plt.ylabel('Number of Incidents')
plt.title('Top 10 Attacker Locations')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Task 15: Visualize Attack Type vs Severity
This stacked bar chart compares severity levels across attack types.

```python
import matplotlib.pyplot as plt

attack_severity = pd.crosstab(df['attack_type'], df['severity'])

plt.figure(figsize=(12, 7))
attack_severity.plot(kind='bar', stacked=True, figsize=(12, 7))
plt.xlabel('Attack Type')
plt.ylabel('Number of Incidents')
plt.title('Attack Type vs Severity')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

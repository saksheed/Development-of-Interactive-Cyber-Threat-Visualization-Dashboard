# SQL Task
Imported dataset

```python
!pip install pandasql
import pandas as pd
from pandasql import sqldf as pysqldf

file_path = '/content/cybersecurity_incidents_500_records.csv'
df = pd.read_csv(file_path)
display(df.head())
```

#### Task 1: Display sample records
This query previews the first 10 records from the dataset.

```python
query = """SELECT * FROM df LIMIT 10"""
display(pysqldf(query))
```

#### Task 2: Count incidents by attacker location
This query counts the number of cybersecurity incidents for each `attacker_location`.

```python
query = """SELECT attacker_location, COUNT(*) AS incident_count FROM df GROUP BY attacker_location ORDER BY incident_count DESC"""
display(pysqldf(query))
```

#### Task 3: Distribution of attack types
This query shows how frequently each attack type occurs.

```python
query = """SELECT attack_type, COUNT(*) AS incident_count FROM df GROUP BY attack_type ORDER BY incident_count DESC"""
display(pysqldf(query))
```

#### Task 4: Severity distribution
This query analyzes how incidents are distributed by severity.

```python
query = """SELECT severity, COUNT(*) AS incident_count FROM df GROUP BY severity ORDER BY incident_count DESC"""
display(pysqldf(query))
```

#### Task 5: Attacks by target device
This query identifies the most targeted devices.

```python
query = """SELECT attack_device, COUNT(*) AS incident_count FROM df GROUP BY attack_device ORDER BY incident_count DESC"""
display(pysqldf(query))
```

#### Task 6: Security controls vs severity
This query analyzes severity distribution across security mechanisms.

```python
query = """SELECT security_type, severity, COUNT(*) AS incident_count FROM df GROUP BY security_type, severity ORDER BY security_type, incident_count DESC"""
display(pysqldf(query))
```

#### Task 7: Classify incidents into risk categories
This query creates logical risk labels using severity.

```python
query = """SELECT
    attacker_location,
    attack_type,
    severity,
    CASE
        WHEN severity = 'Critical' THEN 'Severe Risk'
        WHEN severity = 'High' THEN 'High Risk'
        WHEN severity = 'Medium' THEN 'Moderate Risk'
        ELSE 'Low Risk'
    END AS risk_category
FROM df"""
display(pysqldf(query))
```
